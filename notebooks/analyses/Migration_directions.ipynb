```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import numpy as np
```

```python
df_students_take_tests = pd.read_csv("../tables_creation/final_tables/students_take_tests.csv")
df_students = pd.read_csv("../tables_creation/final_tables/students.csv")
df_location = pd.read_csv('../tables_creation/final_tables/locations.csv')
```

```python
df_students_take_tests = df_students_take_tests.loc[(df_students_take_tests['test_subject'] == 'ukrtest') & (df_students_take_tests['year'] == 2022)]
df_students_take_tests
```

```python
df_merged = df_students_take_tests.merge(df_students, how='left', on='outid')
df_merged = df_merged.loc[df_merged['status'] == 'A graduate of an Ukrainian school of the current year']
df_merged = df_merged.loc[:, ['outid', 'test_status', 'status', 'KATOTTG_2023_test_center', 'KATOTTG_2023']]
df_merged
```

```python
df_merged = df_merged.merge(df_location, how='left', on='KATOTTG_2023')
df_merged.rename(columns={'region_name': 'region_student'}, inplace=True)
df_merged = df_merged.merge(df_location, how='left', left_on='KATOTTG_2023_test_center', right_on='KATOTTG_2023')
df_merged.rename(columns={'region_name': 'region_test_center'}, inplace=True)
df_merged = df_merged.loc[:, ['outid', 'test_status', 'region_student', 'category_y', 'region_test_center']]
```

```python
df_merged['is_migrated'] = (df_merged['region_student'] != df_merged['region_test_center'])
```
