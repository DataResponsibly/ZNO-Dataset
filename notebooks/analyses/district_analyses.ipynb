```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

```python
df_st_take_test = pd.read_csv('../tables_creation/final_tables/students_take_tests.csv', dtype={'KOATUU_2020_test_center': str})
# not na in year
df_st_take_test = df_st_take_test[df_st_take_test['year'].notna()]
df_st_take_test.head()
```

```python
df_location = pd.read_csv('../tables_creation/final_tables/locations.csv', dtype={'KOATUU_2020': str})
df_location.head()
```

```python
df_location_centers = df_location.rename(columns = {'KOATUU_2020':'KOATUU_2020_test_center', 'region_name':'region_name_test_center', 'KATOTTG_2023':'KATOTTG_2023_test_center', 'category':'category_test_center'})
df_location_centers
```

```python
# join by KOATUU_2020
df_st_take_test = pd.merge(df_st_take_test, df_location_centers[['region_name_test_center','KOATUU_2020_test_center','KATOTTG_2023_test_center']], on='KATOTTG_2023_test_center', how='inner')[['outid','year', 'region_name_test_center', 'KOATUU_2020_test_center','KATOTTG_2023_test_center']]
df_st_take_test
```

```python
# get onyl 2021 and 2022
df_st_take_test = df_st_take_test[df_st_take_test['year'].isin([2021,2022])]
df_st_take_test
```

```python
# if df_st_take_test KOATUU_2020 ends with 00000 it is district
# get only districts
df_st_take_test = df_st_take_test[df_st_take_test['KOATUU_2020_test_center'].str.endswith('00000')]
df_st_take_test
```

```python

df_location_base = pd.read_csv('../tables_creation/final_tables/locations_base.csv', dtype={'KOATUU_2020': str})
# get only district
df_location_base = df_location_base[df_location_base['category'] == 'district']
df_location_base.head()
```

```python
# inner join by KATOTTG_2023
df_st_take_test = pd.merge(df_st_take_test, df_location_base[['KATOTTG_2023','ukrainian_name','english_name']], left_on = "KOATUU_2023_test_center", right_on='KATOTTG_2023', how='inner')[['outid','year', 'region_name_test_center','KATOTTG_2023_test_center', 'ukrainian_name', 'english_name']]
df_st_take_test
```
