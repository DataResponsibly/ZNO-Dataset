```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

```python
df_st_take_test = pd.read_csv('../tables_creation/final_tables/students_take_tests.csv', dtype={'KOATUU_2020_test_center': str})
# not na in year
df_st_take_test = df_st_take_test[df_st_take_test['year'].notna()]
df_st_take_test.head()
```

```python
df_location = pd.read_csv('../tables_creation/final_tables/locations.csv', dtype={'KOATUU_2020': str})
df_location.head()
```

```python
df_location_centers = df_location.rename(columns = {'KOATUU_2020':'KOATUU_2020_test_center', 'region_name':'region_name_test_center', 'KATOTTG_2023':'KATOTTG_2023_test_center', 'category':'category_test_center'})
df_location_centers
```

```python
# join by KOATUU_2020
df_st_take_test = pd.merge(df_st_take_test, df_location_centers[['region_name_test_center','KOATUU_2020_test_center','KATOTTG_2023_test_center']], on='KATOTTG_2023_test_center', how='inner')[['outid','year', 'region_name_test_center', 'KOATUU_2020_test_center','KATOTTG_2023_test_center']]
df_st_take_test
```

```python
# unique students
df_st_take_test = df_st_take_test.groupby(['outid', 'year']).agg({'region_name_test_center': 'first', 'KOATUU_2020_test_center': 'first', 'KATOTTG_2023_test_center':'first'}).reset_index()
df_st_take_test = df_st_take_test.drop_duplicates()
```

```python
# get onyl 2021 and 2022
df_st_take_test = df_st_take_test[df_st_take_test['year'].isin([2021,2022])]
df_st_take_test
```

```python
# if df_st_take_test KOATUU_2020 ends with 00000 it is district
# create district code
df_st_take_test['district_code'] = df_st_take_test['KOATUU_2020_test_center'].apply(lambda x: x[:5] +"00000" if not x.endswith('00000') else x)
df_st_take_test
```

```python

df_location_base = pd.read_csv('../tables_creation/final_tables/locations_base.csv', dtype={'KOATUU_2020': str})
# get only district
#df_location_base = df_location_base[df_location_base['category'] == 'district']
df_location_base.head()
```

```python
df_location_base['KOATUU_2020'].value_counts()
```

```python
# inner join by KATOTTG_2023
df_st_take_test = pd.merge(df_st_take_test, df_location_base[['KOATUU_2020','ukrainian_name','english_name']], left_on = "district_code", right_on='KOATUU_2020', how='inner')[['outid','year', 'region_name_test_center','KATOTTG_2023_test_center', 'ukrainian_name', 'english_name']]
df_st_take_test
```

```python
# unique students
df_st_take_test = df_st_take_test.groupby(['outid', 'year']).agg({'region_name_test_center': 'first', 'KATOTTG_2023_test_center': 'first', 'ukrainian_name':'first', 'english_name': 'first'}).reset_index()
df_st_take_test = df_st_take_test.drop_duplicates()
```

```python
# region without data in 2022
reg_without_data = ['Donetska', 'Khersonska', 'Luhanska']
df_st_take_test = df_st_take_test[~df_st_take_test['region_name_test_center'].isin(reg_without_data)]
```

```python
# create pivot table as matrix (columns - years, rows - regions)
pivot_matrix = df_st_take_test[df_st_take_test.KATOTTG_2023_test_center.str.startswith("UA")].pivot_table(index=['region_name_test_center', 'ukrainian_name', 'english_name'], columns='year', values='outid', aggfunc='count').reset_index()
pivot_matrix
```

```python
# percenage change from 2021 to 2022
pivot_matrix['percentage_change'] = ((pivot_matrix[2022] - pivot_matrix[2021])/pivot_matrix[2021])*100
pivot_matrix
```

```python
# if none in 2021, then 100% change
# if none in 2022, then -100% change
def change_percentage(x):
    if pd.isna(x[2021]) and not pd.isna(x[2022]):
        return 100
    elif pd.isna(x[2022]) and not pd.isna(x[2021]):
        return -100
    else:
        return x['percentage_change']
pivot_matrix['percentage_change'] = pivot_matrix.apply(change_percentage, axis=1)
pivot_matrix
```

```python
pivot_matrix = pivot_matrix.sort_values(by='percentage_change', ascending=False)
pivot_matrix
```

```python
pivot_matrix.to_csv('pivot_table_students_take_tests.csv', index=False)
```
