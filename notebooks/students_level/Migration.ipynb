```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
from dictionaries_translation import *
```

```python
zno_translated = pd.read_csv('zno_translated.csv')
zno_translated
```

```python
df_nmt = pd.read_csv('Odata2022File.csv', sep=';')
df_nmt["RegName"] = df_nmt["RegName"].replace(region_dict)
df_nmt["EORegName"] = df_nmt["EORegName"].replace(region_dict)
df_nmt["PTRegName"] = df_nmt["PTRegName"].replace(region_dict)
df_nmt
```

```python
sns.set(style='darkgrid')
```

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.countplot(ax = ax, x=zno_translated["regname"], order=zno_translated["regname"].value_counts().index)
values = ax.bar_label(ax.containers[0])
ax.tick_params(axis='x', rotation=90)
```

Distribution of regions of the participants in 2018-2021.

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.countplot(ax = ax, x=df_nmt["RegName"], order=df_nmt["RegName"].value_counts().index)
values = ax.bar_label(ax.containers[0])
ax.tick_params(axis='x', rotation=90)
```

Distribution of regions of the participants in 2022. The distributions are quite similar.

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.countplot(ax = ax, x=zno_translated["ukrptregname"], order=zno_translated["ukrptregname"].value_counts().index)
values = ax.bar_label(ax.containers[0])
ax.tick_params(axis='x', rotation=90)
```

Distribution of regions of the test centers for Ukrainian language and literature test in 2018-2021.

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.countplot(ax = ax, x=df_nmt["PTRegName"], order=df_nmt["PTRegName"].value_counts().index)
values = ax.bar_label(ax.containers[0])
ax.tick_params(axis='x', rotation=90)
```

Distribution of regions of the test centers for general test in 2022. Distributions are very different: we can see that the label with most of the participants is "Other countries", and Donetsk, Luhansk and Kherson oblasts are not in the list.

```python
regions = []
other_countries = []
migration = []
for region in df_nmt['RegName'].unique():
    regions.append(region)
    other_countries.append(df_nmt[(df_nmt['RegName'] == region) & (df_nmt['PTRegName'] == 'Other countries')].shape[0]/df_nmt[df_nmt['RegName'] == region].shape[0])
    migration.append(df_nmt[(df_nmt['RegName'] == region) & (df_nmt['PTRegName'] != region)].shape[0]/df_nmt[df_nmt['RegName'] == region].shape[0])
    
df_migration = pd.DataFrame({'region': regions, 'other_countries': other_countries, 'migration': migration}).sort_values(by=['migration'], ascending=False)
df_migration
```

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.barplot(ax = ax, x=df_migration["region"], y=df_migration["other_countries"], order=df_migration.sort_values(by=['other_countries'], ascending=False)["region"])
values = ax.bar_label(ax.containers[0], labels=round(df_migration.sort_values(by=['other_countries'], ascending=False)["other_countries"], 2))
ax.tick_params(axis='x', rotation=90)
```

Distribution of the regions by percentage of migration to the other countries.

```python
figure_dims = (15, 6)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.barplot(ax = ax, x=df_migration["region"], y=df_migration["migration"], order=df_migration.sort_values(by=['migration'], ascending=False)["region"])
values = ax.bar_label(ax.containers[0], labels=round(df_migration.sort_values(by=['migration'], ascending=False)["migration"], 2))
ax.tick_params(axis='x', rotation=90)
```

Distribution of the regions by percentage of migration (both inner and to the other countries).
