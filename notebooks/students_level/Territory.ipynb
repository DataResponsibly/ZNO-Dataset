# Matching regions

```python
codifiers['reg_name'] = codifiers['NU'].str.split('/').str[0].str.lower()
codifiers[codifiers.area_code == '000'][['NU', 'region_code', 'reg_name']]
```

Drop '01' and '85' since we don't have them in our dataset

```python
codifiers = codifiers[~codifiers['region_code'].isin(['01', '85'])]
```

```python
codifiers = codifiers[~codifiers['reg_name'].str.startswith('міста ')]
codifiers = codifiers[~codifiers['reg_name'].str.startswith('райони ')]
codifiers = codifiers[~codifiers['reg_name'].str.startswith('селища ')]
codifiers = codifiers[~codifiers['reg_name'].str.startswith('населенi ')]
codifiers = codifiers[~codifiers['reg_name'].str.startswith('селища, ')]
codifiers = codifiers[~codifiers['reg_name'].str.startswith('сільради ')]
codifiers = codifiers.reset_index(drop=True)
```

TE - a codifier in KOATUU
NP - category of the object, where С - село, Щ - селище, Т - селище міського типу,                 М - місто, Р - район міста;
NU - name in Ukrainian.

```python
codifiers.NP.unique()
```

```python
codifiers.loc[codifiers['NP'] == 'М', 'reg_name'] = 'м.' + codifiers.loc[codifiers['NP'] == 'М', 'reg_name']
codifiers.loc[codifiers['NP'] == 'С', 'reg_name'] = 'с.' + codifiers.loc[codifiers['NP'] == 'С', 'reg_name']
codifiers.loc[codifiers['NP'] == 'Щ', 'reg_name'] = 'с-ще ' + codifiers.loc[codifiers['NP'] == 'Щ', 'reg_name']
codifiers.loc[codifiers['NP'] == 'Т', 'reg_name'] = 'смт ' + codifiers.loc[codifiers['NP'] == 'Т', 'reg_name']
```

```python
def change_area_code(place, code, dataset):
    region_code = code[:2]
    dataset.loc[(dataset.region_code == region_code)&(dataset['tername_new'] == place), 'code'] = code
    dataset.loc[(dataset.region_code == region_code)&(dataset['tername_new'] == place), 'area_code'] = code[2:]
    dataset.loc[(dataset.region_code == region_code)&(dataset['tername_new'] == place), 'areaname_new'] = dataset.loc[(dataset.code == code), 'areaname_new'].unique()[0]
```
