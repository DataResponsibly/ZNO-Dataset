```python
languages = {"hung_lang":'Hungarian', "ukr_lang":'Ukrainian', "rus_lang":'Russian', 
        "roman_lang":'Romanian', "pol_lang":'Polish', "mold_lang":'Moldovan', "other_lang":'Other'}

for lang in languages:
    orig_df[lang] = (orig_df["classlangname"] == languages[lang]).astype(int)
```

```python
orig_df.ukrteststatus.unique()
```

```python
def change_teststatus(test):
    orig_df[f"{test}_accept"] = (orig_df[f"{test}teststatus"] == 'Test passed').astype(int)
    orig_df[f"{test}_fail"] = (orig_df[f"{test}teststatus"] == "The score threshold wasn't reached").astype(int)
    orig_df[f"{test}_abscent"] = (orig_df[f"{test}teststatus"] == 'No attendance').astype(int)
    orig_df[f"{test}_annul"] = (orig_df[f"{test}teststatus"] == 'Test results were cancelled').astype(int)
    orig_df[f"{test}_no_choice"] = (orig_df[f"{test}teststatus"] == "100-200 wasn't selected").astype(int)
```

```python
subjects = ["ukr", "hist", "math", "phys", "chem", "bio", "geo", "eng", "fra", "deu", "spa", 'rus']

for sub in subjects:
    change_teststatus(sub)
```

```python
orig_df.mathlang.unique()
```

```python
def change_testlang(test):
    orig_df[f"{test}_ukrlang"] = (orig_df[f"{test}lang"] == 'Ukrainian').astype(int)
    orig_df[f"{test}_ruslang"] = (orig_df[f"{test}lang"] == 'Russian').astype(int)
    orig_df[f"{test}_moldlang"] = (orig_df[f"{test}lang"] == 'Romanian').astype(int)
    orig_df[f"{test}_hunglang"] = (orig_df[f"{test}lang"] == 'Hungarian').astype(int)
    orig_df[f"{test}_romlang"] = (orig_df[f"{test}lang"] == 'Moldovan').astype(int)
    orig_df[f"{test}_pollang"] = (orig_df[f"{test}lang"] == 'Polish').astype(int)
```

```python
subj = ["hist", "math", "phys", "chem", "bio", "geo"]

for sub in subj:
    change_testlang(sub)
```

```python
orig_df = orig_df.drop(['sextypename'], axis=1)
```

```python
to_remove = ['classprofilename', 'classlangname', 'ukrteststatus', 'histteststatus',
                        'mathteststatus', 'physteststatus', 'chemteststatus', 'bioteststatus',
                        'geoteststatus', 'engteststatus', 'frateststatus', 'deuteststatus',
                        'spateststatus', 'mathlang', 'histlang', 'physlang', 'chemlang',
                        'biolang', 'geolang', 'rusteststatus']
for feature in to_remove:
    orig_df = orig_df.drop([feature], axis=1)
```

```python
def q25(x):
    return x.quantile(0.25)

def q50(x):
    return x.quantile(0.5)

def q75(x):
    return x.quantile(0.75)
```

```python
list(orig_df.columns)
```

```python
orig_df['Degree of institution'].unique()
```

```python
orig_df["degree_I_III"] = (orig_df['Degree of institution'] == 'I-III').astype(int)
orig_df["degree_II_III"] = (orig_df['Degree of institution'] == 'II-III').astype(int)
orig_df["degree_III"] = (orig_df['Degree of institution'] == 'III').astype(int)
orig_df["degree_I"] = (orig_df['Degree of institution'] == 'I').astype(int)
orig_df["degree_I_II"] = (orig_df['Degree of institution'] == 'I-II').astype(int)
```

```python
orig_df = orig_df.drop(['Degree of institution'], axis=1)
```

```python
ag_dct = {col:'sum' for col in orig_df.columns if col not in ['eoname', 'year', 'age']}
finance_dct= {'Number of full-time teaching staff in total':'max',
'Number of full-time non-teaching staff in total':'max', 'Number of full-time teaching staff of retirement age':'max',
 'Number of students in total':'max','Number of classes in total':'max', 
 'Total number of students and children receiving preschool education':'max',
 'Expenditures for the operation of the institution (general fund), UAH thousand (cash expenditures)':'max',
 'Expenditures for the operation of the institution (general fund), UAH thousand (plan)':'max'}

ag_dct.update(finance_dct)

for col in ag_dct:
    if 'ball' in col:
        ag_dct[col] = ['min', 'max', q25, q50, q75, 'mean']
    elif 'test' in col:
        ag_dct[col] = 'count'
    elif '_lang' in col:
        ag_dct[col] = 'max'
    elif 'degree' in col:
        ag_dct[col] = 'max'
    elif col.startswith('eo'):
        ag_dct[col] = 'min'

ag_dct['ukradaptscale'] = 'first'
```

```python
ag_dct
```

```python
grouped = orig_df.groupby(["eoname", "year"],dropna=True).agg(ag_dct)
print("SUCCESS")
```

```python
grouped = grouped.reset_index()
```
