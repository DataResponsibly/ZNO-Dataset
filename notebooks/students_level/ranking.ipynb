```python
import pandas as pd
```

```python
df = pd.read_csv('eie_full_translated.csv')
```

Choose results for 2020 

```python
df_2020 = df[df.year == 2020]
df_2020
```

Choose students who chose and passed four subjects: Ukrainian, History, Mathematics and English

```python
df_four = df_2020[(df_2020['ukrteststatus'] == 'Test passed')&(df_2020['histteststatus']== 'Test passed')&(df_2020['mathteststatus'] == 'Test passed')&(df_2020['engteststatus'] == 'Test passed')]
```

```python
df_four_scores = df_four[['sextypename','ukrball100', 'histball100', 'engball100', 'mathball100']]
df_four_scores
```

On the early stage of EIE in Ukraine it was used the sum of scores for the ranking task for admission. Every participant could choose up to 4 subjects. Ukrainian language and literature always was obligatory to take. Usually it was enough 3 subjects to apply for university. However, a lot of specializations have different requirements that's why it was common practice to choose 4 to have larger variety of specializations to choose.

For example, to apply for Computer Science usually it is needed 3 subjects: Ukrainian language and literature, Mathematics, and English language.

For English Philology it is needed 3 subjects: Ukrainian language and literature, History of Ukraine, and English language.

Let in 2020 be particpants who decided to take the corresponding 4 subjects and we will investigate their chances to be admitted.


```python
df_four_scores['mean_ukr_math_eng'] = round((df_four_scores.ukrball100+df_four_scores.mathball100+df_four_scores.engball100)/3,2)
df_four_scores['mean_ukr_hist_eng'] = round((df_four_scores.ukrball100+df_four_scores.histball100+df_four_scores.engball100)/3,2)
```

```python
display(df_four_scores[['sextypename', 'mean_ukr_hist_eng']].sort_values('mean_ukr_hist_eng', ascending=False).head(7))
display(df_four_scores[['sextypename','mean_ukr_math_eng']].sort_values('mean_ukr_math_eng', ascending=False).head(7))
```

In the last several years the universities in Ukraine define their own weighted coefficients for the coressponding subjects. For example, one of the best universities in Ukraine Taras Shevchenko National University of Kyiv in 2020 used the following weights:

Computer Science: Ukrainian language and literature (0.2), Mathematics (0.6), and History of Ukraine or English language (0.2).

English Philology: Ukrainian language and literature (0.3), History of Ukraine or Mathematics (0.2), and English language (0.5).

Let a participant chose only 3 subjects Ukrainian language and literature, Mathematics and English language. He/she could apply for both specializations. Let's calculate chances

```python
df_four_scores['weights_CS'] = df_four_scores.ukrball100*0.2+df_four_scores.mathball100*0.6+df_four_scores.engball100*0.2
df_four_scores['weights_EngPhil'] = df_four_scores.ukrball100*0.3+df_four_scores.mathball100*0.2+df_four_scores.engball100*0.5
```

```python
display(df_four_scores[['sextypename','weights_CS']].sort_values('weights_CS', ascending=False).head(7))
display(df_four_scores[['sextypename','weights_EngPhil']].sort_values('weights_EngPhil', ascending=False).head(7))
```
