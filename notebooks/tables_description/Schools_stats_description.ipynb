Features:
- EDRPOU (unique code of the legal entity or physical person): 9732 unique codes of schools.
- KOATUU_2020_reg (region level of KOATUU code): 25 unique codes denoting regions of school locations (the first two digits of the full KAOTUU code of the location).
- eotype (type of school's settlement). Range of values:
    - rural
    - urban
- eolevel (level of the school; I is elementary school, II is middle school, III is high school; only schools which have III of education have graduates who want to enter universities, and take tests). Range of values:
    - I-III
    - II-III
    - III
- teachstuff (number of teaching stuff). Range of values:
    - 2-215 (integer)
- nonteachstuff (number of non-teaching stuff). Range of values:
    - 0-165 (integer)
- teachstuffretage (number of teaching stuff of retirement age). Range of values:
    - 0-54 (integer)
- pupils (number of pupils). Range of values:
    - 5-2477 (integer)
- classes (number of classes). Range of values:
    - 0-98 (integer)
- opex (expenses for the operation of the school, current year, thousands UAH). Range of values:
    - 247.787-48330.2 (float)
- opexplan (planned expenses for the operation of the school, next year, thousands UAH). Range of values:
    - 0: no information about planned expanses
    - 56.794-83878.7 (float)
- hub (is the school a hub school). Range of values:
    - N/A: no info
    - yes
    - no
- year (year of the statistics info for the school). Range of values:
    - 2019-2020 (integer)
