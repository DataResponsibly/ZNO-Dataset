# Description of `Students` table

```python
import pandas as pd
import seaborn as sns
sns.set_theme()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../tables_creation/data/Students.csv')
df['EDRPOU'] = df['EDRPOU']
df.head()
```

```python
df.info()
```

```python
# print unique values for each column
for col in df.columns:
    print('Unique length in ', col+':\t', len(df[col].unique()))
```

```python
plt.figure(figsize=(15,8))
plot = sns.countplot(df, x="birth")
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
None
```

```python
plot = sns.countplot(df, x="sextypename")
None
```

```python
df['sextypename'].value_counts()
```

```python
plot = sns.countplot(df, x="classprofilename", order = df['classprofilename'].value_counts().index)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
None
```

```python
df['classprofilename'].value_counts()
```

```python
plot = sns.countplot(df, x="classlangname", order = df['classlangname'].value_counts().index)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
None
```

```python
df['classlangname'].value_counts()
```

Features:
- outid (Student ID): 2490052 unique students identifiers
- birth: year of student's birth. Range of values:
    - 1944-2010 (integer)
- sextypename: sex of student. Range of values:
    - чоловіча: male
    - жіноча: female
- classprofilename: class profile. Range of values:
    - Універсальний: Universal
    - Молодший спеціаліст: Junior Specialist
    - Української філології: Ukrainian Philology
    - Кваліфікований робітник: Qualified Worker
    - Іноземної філології: Foreign philology
    - Математичний: Mathematical
    - Технологічний: Technological
    - Історичний: Historical
    - Інший(багатопрофільність): Other (multi-profile)
    - Інформаційно-технологічний: Information technology
    - Інший: Other
    - Біолого-хімічний: Biological and chemical
    - Фізико-математичний: Physical and mathematical
    - Економічний: Economic
    - Правовий: Legal
    - Військово-спортивний: Military and sports
    - Спортивний: Sports
    - Біотехнологічний: Biotechnological
    - Екологічний: Ecological
    - Біологічний: Biological
    - Географічний: Geographical
    - Художньо-естетичний: Artistic and aesthetic
    - Фізичний: Physical
    - Хіміко-технологічний та агрохімічний: Chemical-technological and agrochemical
    - Біолого-фізичний: Biological and physical
    - Філософський: Philosophical
    - Фізико-хімічний: Physico-chemical
- classlangname: language of class. Range of values:
    - українська: Ukrainian
    - російська: russian
    - румунська: Romanian
    - угорська: Hungarian
    - молдовська: Moldovan
    - польська: Polish
    - інша: Other
- KOATUU_2020 (KOATUU code): 18470 unique codes, denoting schools locations (a unique digital code for each administrative-territorial entity of Ukraine, used from January 1998 till January 2021)
- EDRPOU (unique code of the legal entity or physical person): 21788 unique codes of schools.
