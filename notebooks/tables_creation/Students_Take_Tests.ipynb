```python
for year, dataset in datasets.items():
    #lowercase all columns and add year as attribute
    dataset.columns = [col.lower() for col in dataset.columns]
    dataset['year'] = year

    if year == 2016:
        dataset = dataset.rename(columns=eie_2016)

    if year == 2022:
        dataset = dataset.rename(columns=nmt_2022)
        for type_ in ['ukr', 'hist', 'math']:
            dataset.loc[:, f'{type_}ptregname'] = dataset['ptregname']
            dataset.loc[:, f'{type_}ptareaname'] = dataset['ptareaname']
            dataset.loc[:, f'{type_}pttername'] = dataset['pttername']
            dataset.loc[:, f'{type_}teststatus'] = dataset['teststatus']
        dataset.drop(columns=['ptregname', 'ptareaname', 'pttername','teststatus'], inplace=True)
    elif year == 2023:
        dataset = dataset.rename(columns=nmt_2023)
        for type_ in ['ukr', 'hist', 'math', 'phys', 'chem', 'bio', 'eng', 'fra', 'deu', 'spa']:
            dataset.loc[:, f'{type_}ptregname'] = dataset['ptregname']
            dataset.loc[:, f'{type_}ptareaname'] = dataset['ptareaname']
            dataset.loc[:, f'{type_}pttername'] = dataset['pttername']
        dataset.drop(columns=['ptregname', 'ptareaname', 'pttername'], inplace=True)
    dataset.reset_index(drop=True, inplace=True)
    datasets.update({year:dataset})
```

```python
from itertools import cycle
import pandas as pd


def is_scalar(obj):
    if isinstance(obj, str):
        return True
    elif hasattr(obj, "__iter__"):
        return False
    else:
        return True


def multi_melt(
    df: pd.DataFrame,
    id_vars=None,
    value_vars=None,
    var_name=None,
    value_name="value",
    col_level=None,
    ignore_index=True,
) -> pd.DataFrame:

    # Note: we don't broadcast value_vars ... that would seem unintuitive
    value_vars = value_vars if not is_scalar(value_vars[0]) else [value_vars]
    var_name = var_name if not is_scalar(var_name) else cycle([var_name])
    value_name = value_name if not is_scalar(value_name) else cycle([value_name])

    melted_dfs = [
        (
            df.melt(
                id_vars,
                *melt_args,
                col_level,
                ignore_index,
            ).pipe(lambda df: df.set_index([*id_vars, df.groupby(id_vars).cumcount()]))
        )
        for melt_args in zip(value_vars, var_name, value_name)
    ]

    return (
        pd.concat(melted_dfs, axis=1)
        .sort_index(level=2)
        .reset_index(level=2, drop=True)
        .reset_index()
    )
```

```python
df = df.pipe(
    multi_melt,
    id_vars=["outid","year"],
    value_vars=[[col for col in df.columns if 'ball100' in col], 
                [col for col in df.columns if 'ball12' in col],
                [col for col in df.columns if col.endswith('ball')],
                [col for col in df.columns if col.endswith('teststatus')],
                [col for col in df.columns if col.endswith('ptname')],
                [col for col in df.columns if col.endswith('ptregname')],
                [col for col in df.columns if col.endswith('ptareaname')],
                [col for col in df.columns if col.endswith('pttername')],
                ],
    var_name=["score100_id", "score12_id", "score_id", "test_status_id", "testcenter_name_id",
                 "test_reg_id", "test_area_id", "test_ter_id"],
    value_name=["score100", "score12", "score", "test_status_", "testcenter_name", "test_reg", "test_area", "test_ter"],
)
```
