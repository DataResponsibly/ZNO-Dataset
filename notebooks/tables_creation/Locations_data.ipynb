```python
category_dict = {'О': 'region', 'Р':'district', 'Н':'hromada', 'С':'village', 'Х':'settlement', 'М':'city', 'Т':'urban village', 'В':'districts in cities', 'К':'capital'}
```

```python
for abbr, name in category_dict.items():
    comparison.loc[comparison.category == abbr, 'category'] = name
```

```python
# load location dictionary to match names
locations = pd.read_csv('./matching_data/locations.csv', encoding='utf-8', dtype=str)
locations
```

```python
#check whether pairs ['KOATUU_2020',	'KATOTTG_2023'] are the same for both dataframe
check_pairs = merged_df.merge(locations[['KOATUU_2020',	'KATOTTG_2023']], on = 'KATOTTG_2023', how = 'left')
check_pairs[(check_pairs.KOATUU_2020_y.notna())&(check_pairs.KOATUU_2020_y!=check_pairs.KOATUU_2020_x)]
```

```python
locations_abroad = locations[['KATOTTG_2023', 'KOATUU_2020', 'category', 'tername']][locations.KATOTTG_2023.str.startswith('OC')]
locations_abroad.drop_duplicates(inplace=True)
locations_abroad
```

```python
locations_abroad.rename(columns={'tername':'name'}, inplace=True)
locations_abroad.columns
```

```python
locations_abroad.name.unique()
```

```python
locations_abroad['english name'] = None
```

```python
for name, english_name in renam.dct_abroad_translation.items():
    locations_abroad.loc[(locations_abroad.name == name), 'english name'] = english_name
```

```python
ukrainian_name = {code[:4]:name for name, code in renam.dct_code_area_2023.items()}
for code, country in renam.dct_code_regions.items():
    if code[0] == 'O':
        new_row = {
            'KATOTTG_2023': code+'0'*15,
            'KOATUU_2020': '00'+code[2:]+'0'*6,
            'category': 'abroad',
            'name': ukrainian_name[code],
            'english name': country
        }
        locations_abroad = pd.concat([locations_abroad, pd.DataFrame([new_row])], ignore_index=True)
```

```python
locations_abroad
```

```python
merged_df = pd.concat([merged_df, locations_abroad], ignore_index=True)
merged_df
```

```python
merged_df.rename(columns={'name':'ukrainian_name', 'english name':'english_name'}, inplace=True)
merged_df
```
