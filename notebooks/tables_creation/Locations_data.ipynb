О – області та АРК;
К – міста, що мають спеціальний статус (Київ та Севастополь);
Р – райони в областях та в АРК;
Н – територіальні громади;
М – міста;
Т – селища міського типу;
С – села;
Х – селища;
В – райони в містах.

```python
category_dict = {'О': 'region', 'Р':'district', 'Н':'hromada', 'С':'village', 'Х':'settlement', 'М':'city', 'Т':'urban village', 'В':'districts in cities', 'К':'capital'}
```

```python
for abbr, name in category_dict.items():
    comparison.loc[comparison.category == abbr, 'category'] = name
```

```python
renaming = pd.read_csv('./location_data/translation_location.csv')
renaming
```

```python
renaming.rename(columns = {'ukrainian name':'name'}, inplace=True)
renaming
```

```python
merged_df = comparison.merge(renaming, on='name', how='left')
merged_df
```
