# 1. Demographics, education institutions and general features

## Year of birth and age

```python
figure_dims = (9, 7)
fig, ax = pyplot.subplots(figsize=figure_dims)

sns.countplot(ax = ax, x=zno_translated["year"])
values = ax.bar_label(ax.containers[0])
```
