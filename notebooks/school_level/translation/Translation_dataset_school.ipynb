```python
import pandas as pd
import json
from deep_translator import GoogleTranslator
```

```python
zno_orig = pd.read_csv('school_dataset_sample.csv')
translate_features = ['eoname', 'eotypename_min', 'eoregname_min',
                      'eoareaname_min', 'eotername_min', 'eoparent_min']
translator_google = GoogleTranslator(source='uk', target='en')
```

```python
zno_orig.head()
```

```python
zno_orig.info()
```

```python
zno_orig.columns.to_series().groupby(zno_orig.dtypes).groups
```

```python
column_names = list(zno_orig.columns)
column_names_ukr = column_names[282:]
```

```python
translated_column_names = {}
for name in column_names_ukr:
    translated_column_names[name] = translator_google.translate(name)
```

```python
translated_column_names
```

```python
zno_orig = zno_orig.rename(columns=translated_column_names)
zno_orig.head()
```

```python
translated_columns = []
for feature in translate_features:
    to_tr = zno_orig[feature].unique()
    print("translating", feature, len(to_tr))
    translated_google = {}
    for el in to_tr:
        if not pd.isnull(el):
            translated_google[el] = translator_google.translate(el)
    translated_columns.append(translated_google)
```

```python
len(translated_columns)
```

```python
for feat_i in range(len(translate_features)):
    print("translating", translate_features[feat_i], len(zno_orig[translate_features[feat_i]].unique()))
    for i in range(0, len(translated_columns[feat_i]) + 1, 1000):
        tr_now = dict(list(translated_columns[feat_i].items())[i:i+1000])
        zno_orig.replace({translate_features[feat_i]: tr_now}, inplace=True)
    print(len(zno_orig[translate_features[feat_i]].unique()))
```
